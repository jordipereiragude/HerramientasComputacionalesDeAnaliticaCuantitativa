```julia
# N-Queens

global counter=0

function nextQueen(current,n,posiciones)
    if current==n+1
        global counter += 1
        #println(counter," ",posiciones)
        return
    end
    for test in 1:n
        wrong=false
        for i in 1:current-1
            if posiciones[i]==test
                wrong=true
                break
            end
            if posiciones[i]+(current-i)==test
                wrong=true
                break
            end
            if posiciones[i]-(current-i)==test
                wrong=true
                break
            end
        end
        if wrong==false
            posiciones[current]=test
            nextQueen(current+1,n,posiciones)
        end
    end
    return
end

n=15
#pos=Array{Int64}(n)
pos=zeros(Int64,n)
@time(nextQueen(1,n,pos))
```

```julia
# N-Queens Alternativo

global counter=0

function nextQueenAlt(current,n,posiciones)
    if current==n+1
        global counter += 1
        print(counter,"\t")
        for i in 1:n
            for j in 1:n
                if posiciones[i,j]==(-1)
                    print(j," ")
                end
            end
        end
        print("\n")
        return
    end
    for test in 1:n
        #println(current," ",test)
        if posiciones[current,test]==0
            posiciones[current,test]= -1
            #deberíamos tener un número de filas con 0 en cada columna
            for i in current+1:n
                posiciones[i,test] += 1
            end
            for i in current+1:n
                if test+(i-current)<=n
                    posiciones[i,test+(i-current)] += 1
                    #si el valor es 1, entonces hay que eliminar una fila de la columna i
                end                    
                if test-(i-current)>=1
                    posiciones[i,test-(i-current)] += 1
                    #si el valor es 1, entonces hay que eliminar una fila de la columna i
                end                    
            end
            #if todas las columnas siguientes tienen valor >= 1
            nextQueenAlt(current+1,n,posiciones)
            
            #en las restas, si algo pasa a 0 sumo una fila de la columna i
            for i in current+1:n
                posiciones[i,test] -= 1
            end
            for i in current+1:n
                if test+(i-current)<=n
                    posiciones[i,test+(i-current)] -= 1
                end                    
                if test-(i-current)>=1
                    posiciones[i,test-(i-current)] -= 1
                end                    
            end            
            posiciones[current,test]=0
        end
    end
    return
end

n=15
pos=Array{Int64}(n,n)
for i in 1:n
    for j in 1:n
        pos[i,j]=0
    end
end
@time(nextQueenAlt(1,n,pos))
```
